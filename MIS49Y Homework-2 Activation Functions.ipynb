# Activation Functions

The activation function decides whether a neuron should be activated or not by calculating the weighted sum and further adding bias to it.

```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

# linear activation function

The linear activation function, is where the activation is proportional to the input. The function doesn't do anything to the weighted sum of the input, it simply spits out the value it was given.

```python
def linear_activation(x, m, b):
    return m * x + b
```

```python
x_values = np.linspace(-100, 300, 400)

# Setting different slope and y-intercept values
slope_values = [1, -0.5, 2]
intercept_values = [0, 2, -1]

plt.figure(figsize=(10, 6))

for m, b in zip(slope_values, intercept_values):
    y_values = linear_activation(x_values, m, b)
    plt.plot(x_values, y_values, label=f'm={m}, b={b}')

plt.title('Linear Activation Functions')
plt.xlabel('Input (x)')
plt.ylabel('Output (f(x))')
plt.legend()
plt.grid(True)
plt.show()
```

# binary step function

Binary Step is the simplest threshold-based activation function which works by either activating the node, i.e. passing on its output to the next layer in case the threshold value is surpassed, or doing nothing at all. If the node input value is less than 0, it returns 0 as output. Else, it returns 1.

```python
def binary_step(x, threshold=0):
    return np.where(x > threshold, 1, 0)
```

```python
x_values = np.linspace(-5, 5, 100)

binary_step_output = binary_step(x_values)

# Plot the results
plt.plot(x_values, binary_step_output, label='Binary Step')
plt.title('Binary Step Activation Function')
plt.xlabel('Input')
plt.ylabel('Output')
plt.grid(True)
plt.show()
```

# sigmoid activation function

Sigmoid function exists between (0 to 1). Therefore, it is especially used for models where we have to predict the probability as an output. Since probability of anything exists only between the range of 0 and 1, sigmoid is the right choice.

```python
def sigmoid_activation(x):
    return 1 / (1 + np.exp(-x))
```

```python
x_values = np.linspace(-20, 30, 200)

y_values = sigmoid_activation(x_values)

plt.figure(figsize=(8, 5))
plt.plot(x_values, y_values, label='Sigmoid Activation')
plt.title('Sigmoid Activation Function')
plt.xlabel('Input (x)')
plt.ylabel('Output (f(x))')
plt.legend()
plt.grid(True)
plt.show()
```

# tanh activation function

```python
def tanh_activation(x):
    return np.tanh(x)
```

```python
# Generate input values
x_values = np.linspace(-10, 10, 100)

y_values = tanh_activation(x_values)

plt.figure(figsize=(8, 5))
plt.plot(x_values, y_values, label='Tanh Activation')
plt.title('Tanh Activation Function')
plt.xlabel('Input (x)')
plt.ylabel('Output (f(x))')
plt.legend()
plt.grid(True)
plt.show()
```

As an example, we are calculating whether a person will buy an insurance from an insurance company or not based on their age. We are giving 18, 60 and 100 as age inputs.

```python
# Generate some example data (age)
ages = np.linspace(18, 60, 100)

# Use the tanh function to simulate the output
tanh_output = tanh_activation(0.1 * (ages - 30))  # Adjusting parameters for demonstration

# Plot the tanh activation function with example data
plt.figure(figsize=(8, 5))

# Plot tanh function
plt.plot(ages, tanh_output, label='Tanh Activation')

# Add labels and legend
plt.title('Tanh Activation for Insurance Purchase Prediction')
plt.xlabel('Age')
plt.ylabel('Tanh Output')
plt.legend()
plt.grid(True)
plt.show()
```

The limitation of the tanh function is that it suffers from the vanishing gradient problem. This simply means that as the input function becomes very small or very large depending on the case, the gradient of the function approaches zero, thus making it difficult for the network to update the weights of the earlier layers to learn from the input data. This is usually a big problem in deep neural networks having many layers, since the gradients can become extremely small by the time they reach the earlier layers. This leads to slow convergence and poor performance.

```python
x = np.linspace(-5, 5, 100)

# Compute the gradients of the tanh function
tanh_grad = 1 - np.tanh(x)**2

# Plot the gradients of the tanh function
plt.plot(x, tanh_grad, label='tanh')
plt.xlabel('Input')
plt.ylabel('Gradient')
plt.title('Gradients of the tanh activation function')
plt.legend()
plt.show()
```

As we can see from the plot, the gradient of the tanh function approaches zero as the input becomes very large or very small.

# relu activation function

The rectified linear activation function or ReLU for short is a piecewise linear function that will output the input directly if it is positive, otherwise, it will output zero. It's known for its simplicity and effectiveness in promoting sparsity and overcoming the vanishing gradient problem.

```python
def relu_activation(x):
    return np.maximum(0, x)
```

```python
x_values = np.linspace(-5, 25, 200)

y_values = relu_activation(x_values)

plt.figure(figsize=(8, 5))
plt.plot(x_values, y_values, label='ReLU Activation')
plt.title('ReLU Activation Function')
plt.xlabel('Input (x)')
plt.ylabel('Output (f(x))')
plt.legend()
plt.grid(True)
plt.show()
```

# softmax activation function

This function is typically used in the output layer of a neural network for multi-class classification. It converts a vector of numbers into a vector of probabilities, where the probabilities of each value are proportional to the relative scale of each value in the vector.

```python
def softmax(x):
    e_x = np.exp(x - np.max(x))
    return e_x / e_x.sum(axis=0)
```

```python
x = np.linspace(-10, 10, 100)

y_softmax = softmax(x)

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.plot(x, y_softmax, label="Softmax")
plt.title("Softmax Activation Function")
plt.xlabel("Input value")
plt.ylabel("Softmax output")
plt.legend()
```

# leakyRELU activation function

This is a variation of the ReLU (Rectified Linear Unit) function, which allows a small, non-zero gradient when the unit is not active.

```python
def leaky_relu(x):
    return np.where(x > 0, x, x * 0.01)
```

```python
x = np.linspace(-20, 40, 300)

y_leaky_relu = leaky_relu(x)

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 2)
plt.plot(x, y_leaky_relu, label="LeakyReLU")
plt.title("LeakyReLU Activation Function")
plt.xlabel("Input value")
plt.ylabel("LeakyReLU output")
plt.legend()

plt.show()
```
